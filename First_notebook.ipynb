```python
#!pip install nltk

import nltk
nltk.download('popular');
from nltk.corpus import stopwords
from nltk import word_tokenize
import string
from collections import Counter
import numpy as np
from collections import OrderedDict
```

```python
import pandas as pd

resumes = pd.read_csv('resumes.csv')
#resumes.iloc[:, 1].values
```

```python
#### creating dictionary of keys as person name, values as his/her resume

resumes_dict = {}
for i in range(len(resumes)):
    resumes_dict[resumes.iloc[i, 0]] = resumes.iloc[i, 1]
```

```python
def wordList_removePuncs(resumes_dict):
    stop = stopwords.words('english') + list(string.punctuation) + ['\n']
    wordList = []
    for doc in resumes_dict.values():
        for word in word_tokenize(doc.lower().strip()): 
            if not word in stop:
                wordList.append(word)
    return wordList
```

```python
word_list_redundant = wordList_removePuncs(resumes_dict)

vocab = list(set(word_list_redundant))
```

```python
def termFrequencyInDoc(vocab, doc_dict):
    tf_docs = {}
    for doc_id in doc_dict.keys():
        tf_docs[doc_id] = {}
    
    for word in vocab:
        for doc_id,doc in doc_dict.items():
            tf_docs[doc_id][word] = doc.count(word)
    return tf_docs

tf_doc = termFrequencyInDoc(vocab, resumes_dict)
```

```python
tf_doc
```

```python
def wordDocFre(vocab, doc_dict):
    df = {}
    for word in vocab:
        frq = 0
        for doc in doc_dict.values():
#             if word in doc.lower().split():
            if word in word_tokenize(doc.lower().strip()):
                frq = frq + 1
        df[word] = frq
    return df
```

```python
doc_fr = wordDocFre(vocab, resumes_dict)
doc_fr
```

```python
def inverseDocFre(vocab,doc_fre,length):
    idf= {} 
    for word in vocab:     
        idf[word] = np.log2((length+1) / doc_fre[word])
    return idf

```

```python
idf = inverseDocFre(vocab,doc_fr,len(resumes))
idf
```

```python
def tfidf(vocab,tf,idf_scr,doc_dict):
    tf_idf_scr = {}
    for doc_id in doc_dict.keys():
        tf_idf_scr[doc_id] = {}
    for word in vocab:
        for doc_id,doc in doc_dict.items():
            tf_idf_scr[doc_id][word] = tf[doc_id][word] * idf_scr[word]
    return tf_idf_scr
```

```python
tf_idf_scr = tfidf(vocab,tf_doc,idf,resumes_dict)
tf_idf_scr
```

```python
def vectorSpaceModel(query, doc_dict,tfidf_scr):
    query_vocab = []
    for word in query.split():
        if word not in query_vocab:
            query_vocab.append(word)

    query_wc = {}
    for word in query_vocab:
        query_wc[word] = query.lower().split().count(word)
    
    relevance_scores = {}
    for doc_id in doc_dict.keys():
        score = 0
        for word in query_vocab:
            score += query_wc[word] * ((tf_doc[doc_id][word]) * 3) / (tf_doc[doc_id][word] + 2) #tfidf_scr[doc_id][word]
        relevance_scores[doc_id] = score
    sorted_value = OrderedDict(sorted(relevance_scores.items(), key=lambda x: x[1], reverse = True))
    top_5 = {k: sorted_value[k] for k in list(sorted_value)[:len(resumes)]}
    return top_5
```

```python
top = vectorSpaceModel("energy sector \
                        writing processes  instructions\
                        \
                        training end users\
                        large complex projects\
                        experience business analyst\
                        business process development optimization\
                        \
                        stakeholder management", resumes_dict,tf_idf_scr) 
top
```

```python
csv_file = pd.DataFrame.from_dict(top, orient='index')
```
